```python
#!usr/bin/python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import itur
from sklearn.linear_model import LinearRegression
import warnings
import matplotlib
import tkinter

warnings.filterwarnings("ignore")
#matplotlib.use( 'tkagg' )

def getRainfall(state_name,C_state,data_state,R_LATITUDE,R_LONGITUDE):
    data_final = {}
    data_state[state_name]["PER_DAY_AVG"] = data_state[state_name]["ANNUAL"].div(365)
    durations = [1, 2, 4, 12, 24, 48, 96, 141.17, 300]
    columns = data_state[state_name][["YEAR", "PER_DAY_AVG"]]
    # print(columns)
    data_state_DURATION = columns.copy()
    data_state_DURATION["dura_5min"] = data_state[state_name]["PER_DAY_AVG"].div(300)
    data_state_DURATION["dura_10min"] = data_state[state_name]["PER_DAY_AVG"].div(
        141.17
    )
    data_state_DURATION["dura_15min"] = data_state[state_name]["PER_DAY_AVG"].div(96)
    data_state_DURATION["dura_30min"] = data_state[state_name]["PER_DAY_AVG"].div(48)
    data_state_DURATION["dura_1hr"] = data_state[state_name]["PER_DAY_AVG"].div(24)
    data_state_DURATION["dura_2hr"] = data_state[state_name]["PER_DAY_AVG"].div(12)
    data_state_DURATION["dura_6hr"] = data_state[state_name]["PER_DAY_AVG"].div(4)
    data_state_DURATION["dura_12hr"] = data_state[state_name]["PER_DAY_AVG"].div(2)
    data_state_DURATION["dura_24hr"] = data_state[state_name]["PER_DAY_AVG"].div(1)
    # data_state_DURATION
    data_state_IDF = data_state_DURATION.sort_values(by=["dura_30min"], ascending=False)
    data_state_IDF = data_state_IDF.reset_index()
    data_state_IDF["Rank(M)"] = data_state_IDF.index + 1
    data_state_IDF.drop(
        [
            "index",
            "dura_5min",
            "dura_10min",
            "dura_15min",
            "dura_1hr",
            "dura_2hr",
            "dura_6hr",
            "dura_12hr",
            "dura_24hr",
        ],
        axis=1,
        inplace=True,
    )
    N = len(data_state_IDF.axes[0])
    data_state_IDF["EXCEEDENCE_PROBABILITY(P)"] = data_state_IDF["Rank(M)"].div(N + 1)
    data_state_IDF["RETURN_PERIOD(T)"] = 1 / (
        data_state_IDF["EXCEEDENCE_PROBABILITY(P)"]
    )
    print(data_state_IDF)
    
    
    # print(state_name)
    #regressor = LinearRegression()
    #regressor.fit(data_state_IDF[["RETURN_PERIOD(T)"]], data_state_IDF[["PER_DAY_AVG"]])

    #b = regressor.coef_
    #a = regressor.intercept_
    # print(a)
    # print(b)

    C = C_state
    t = 1
    #R2_24 = (a * np.log(2)) + b
    #R2_24 = float(R2_24)
    data_state_IDF["INTENSITY(mm/hr)"] = (
        C * ((data_state_IDF["RETURN_PERIOD(T)"] ** 0.2) / (t ** 0.71)) * data_state_IDF['PER_DAY_AVG'] ** 0.33
    )
    # data_state_IDF
    print(data_state_IDF)

    data_state_IDF["INTENSITY(mm/hr)"] = (data_state_IDF["INTENSITY(mm/hr)"]) / 100

    data_state_IDF["Rp(p)(mm/hr)"] = itur.models.itu837.rainfall_rate(
        R_LATITUDE, R_LONGITUDE, data_state_IDF["INTENSITY(mm/hr)"]
    )
    print(data_state_IDF)
    #data_state_IDF.plot.scatter(x="RETURN_PERIOD(T)", y="PER_DAY_AVG")
    #plt.show()
    #data_state_IDF.plot.scatter(x="PER_DAY_AVG", y="INTENSITY(mm/hr)")
    data_final[state_name] = data_state_IDF
    # data_state_IDF.head()
    return data_final[state_name]


if __name__ == "__main__":
    # print("sys", len(sys.argv))  # temp
    location_csv = "/content/Area_Weighted_Monthly_Seasonal_And_Annual_Rainfall_0 - SUBDIV_Data_Rainfall_1951-2014.csv"

    if True:  # sys.argv ==2
        

        #if sys.argv[1] == "--help":
          #print("type --ra [Rain Attenuation]\ntype --rf [Rain Rain Fall]")
        if True:
          print("Select number from the following state to continue:")
          state = [
              "ANDAMAN & NICOBAR ISLANDS",
              "ARUNACHAL PRADESH",
              "ASSAM & MEGHALAYA",
              "NAGALAND, MANIPUR, MIZORAM,TRIPURA",
              "SUB-HIMALAYAN W BENGAL & SIKKIM",
              "GANGETIC WEST BENGAL",
              "ORISSA",
              "JHARKHAND",
              "BIHAR ",
              "EAST UTTAR PRADESH",
              "WEST UTTAR PRADESH",
              "UTTARANCHAL",
              "HARYANA, DELHI & CHANDIGARH",
              "PUNJAB ",
              "HIMACHAL PRADESH",
              "JAMMU & KASHMIR",
              "WEST RAJASTHAN ",
              "EAST  RAJASTHAN ",
              "WEST MADHYA PRADESH",
              "EAST MADHYA PRADESH",
              "GUJARAT REGION, DADRA & NAGAR HAVELI",
              "SAURASHTRA KUTCH & DIU",
              "KOKAN & GOA",
              "MADHYA MAHARASHTRA",
              "MARATWADA ",
              "VIDARBHA",
              "CHATTISGARH",
              "COSTAL ANDHRA PRADESH",
              "TELENGANA",
              "RAYALSEEMA",
              "TAMIL NADU & PONDICHERRY",
              "COASTAL KARNATAKA",
              "NORTH INTERIOR KARNATAKA",
              "SOUTH INTERIOR KARNATAKA",
              "KERALA",
              "LAKSHADWEEP ",
          ]
          C=[7.1,9.1,9.1,9.1,9.1,9.1,9.1,9.1,9.1,9.1,8.0,8.0,8.0,8.0,8.0,8.0,8.3,7.7,7.7,7.7,8.3,8.3,8.3,8.3]
          for i in range(len(state)):
              print("{}] {}".format(i + 1, state[i]))
          while 1:
              number_chosen = input("Enter State Number: ")
              try:
                  number_chosen = int(number_chosen)
                  if number_chosen > 36 or number_chosen < 0:
                      print("Please enter number in between 1 and 36")
                  else:
                      print("Selected State :",state[number_chosen-1])
                      break
              except ValueError:
                  print("Please enter valid number")
          data = pd.read_csv(location_csv)
          data_state = {}
          for i in state:
              data_state[i] = data[(data["SD_Name"] == i)]
          for i in state:
              data_state[i].fillna(data_state[i].mean(), inplace=True)
          print("1.Predict Rain Fall")
          print("2.Predict Rain Attenuation")
          print("Enter the operation no. You want to perform:")
          # print(getRainfall(state[number_chosen], data_state))  # temprary
          try :
            operation=int(input())
            if operation==1:
                R_LATITUDE=float(input("Enter Latitude (in °N): "))
                R_LONGITUDE=float(input("Enter Longitude (in °E): "))
                R_Fall_Rate=getRainfall(state[number_chosen-1],C[number_chosen-1],data_state,R_LATITUDE,R_LONGITUDE)["Rp(p)(mm/hr)"]
                R_Fall_Rate_Avg=R_Fall_Rate.mean()
                print("Avg Rain Fall Rate(mm/hr): ",R_Fall_Rate_Avg,"mm/hr")
            elif operation==2:
                R_LATITUDE=float(input("Enter Latitude (in °N): "))
                R_LONGITUDE=float(input("Enter Longitude (in °E): "))
                R_ALTITUDE=float(input("Enter Altitude (in KM): "))
                FREQUENCY=float(input("Enter Frequency (in GHz): "))
                ELEVATION=float(input("Enter Elevation (in degree): "))
                POLARIZATION=float(input("Enter Polarization (in degree): "))
                R_Fall_Rate=getRainfall(state[number_chosen-1],C[number_chosen-1],data_state,R_LATITUDE,R_LONGITUDE)["Rp(p)(mm/hr)"].to_numpy()
                R_attenuation=itur.models.itu618.rain_attenuation(R_LATITUDE, R_LONGITUDE, FREQUENCY, ELEVATION, R_ALTITUDE,p=0.01,R001=R_Fall_Rate, tau=POLARIZATION, Ls=None)
                R_att_avg=R_attenuation.mean()
                print("Avg Rain fall Rate (mm/hr): ",R_Fall_Rate.mean(),"mm/hr\nAvg Rain Attenuation: ",R_att_avg)
            #else:
                #print("type --help for help")
          except ValueError:
            print("ERROR Found ",ValueError)
    #else:
        #print("type help --help to get help")
```
